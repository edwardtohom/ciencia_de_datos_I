```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_ = regr.predict(test_x)

print("Error medio absoluto: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Suma residual de los cuadrados (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```

## Análisis de la precisión del modelo.
El resultado de las tres métricas de evaluación del modelo de regresión lineal son los siguientes:

- Error medio absoluto: 23.27. Este resultado indica que, en promedio, las predicciones del modelo difieren en 23.27 unidades de las emisiones de CO2 reales en el conjunto de prueba. 

- Suma residual de los cuadrados (MSE): 914.49. Este resultado indica la magnitud del error cuadrático medio entre las predicciones del modelo y los datos reales de emisiones de CO2 en el conjunto de prueba. 

- R2-score: 0.69. Este resultado indica que el modelo es capaz de explicar el 69% de la variabilidad en los datos de emisiones de CO2 en el conjunto de prueba. En otras palabras, el modelo tiene una buena capacidad de predicción de las emisiones de CO2 a partir del tamaño del motor en el conjunto de prueba.

1. Utilice el algoritmo K-fold cross validation e indique el polinomio que tenga mejor precisión.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# Cargar los datos
df = pd.read_csv('FuelConsumptionCo2.csv')

# Seleccionar las características
X = df[['ENGINESIZE']]
y = df[['CO2EMISSIONS']]
# Definir los grados de polinomio a probar
degrees = [1, 2, 3, 4, 5]

# Configurar el algoritmo K-fold cross-validation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
# Inicializar variables para guardar los resultados
best_degree = 0
best_error = float('inf')

# Ciclo para probar cada grado de polinomio
for degree in degrees:
    errors = []
    
    # Ciclo para entrenar y evaluar el modelo en cada fold
    for train_index, val_index in kf.split(X):
        # Separar los datos en entrenamiento y validación
        X_train, y_train = X.iloc[train_index], y.iloc[train_index]
        X_val, y_val = X.iloc[val_index], y.iloc[val_index]
        
        # Preprocesar los datos para el grado de polinomio actual
        poly = PolynomialFeatures(degree=degree)
        X_train_poly = poly.fit_transform(X_train)
        X_val_poly = poly.transform(X_val)
        
        # Entrenar el modelo de regresión lineal con los datos de entrenamiento
        model = LinearRegression()
        model.fit(X_train_poly, y_train)
        
        # Evaluar el modelo en los datos de validación
        y_pred = model.predict(X_val_poly)
        error = mean_squared_error(y_val, y_pred)
        errors.append(error)
    
    # Calcular el error medio de validación cruzada para el grado de polinomio actual
    cv_error = np.mean(errors)
    
    # Actualizar el mejor grado de polinomio si corresponde
    if cv_error < best_error:
        best_degree = degree
        best_error = cv_error
# Entrenar el mejor modelo de polinomio en todos los datos de entrenamiento
poly = PolynomialFeatures(degree=best_degree)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

# Evaluar el modelo en los datos de prueba
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)
test_error = mean_squared_error(y_test, y_pred)

# Imprimir los resultados
print('Mejor grado de polinomio:', best_degree)
print('Error medio de validación cruzada:', best_error)
print('Error en los datos de prueba:', test_error)
print('Coeficientes:', model.coef_)
print('Intercepto:', model.intercept_)
```

## Analice las variables e implemente un modelo de regresión múltiple que mejore la predicción del modelo.
Para ello, se utilizaron agreraron las variables de cylinders, fuelconsumption_comb y con ello se obtuvo una precisión del 87%

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

data = pd.read_csv("FuelConsumptionCo2.csv")

# Seleccionar las variables de interés
data = data[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]

# Dividir los datos en conjunto de entrenamiento y prueba
train, test = train_test_split(data, test_size=0.2, random_state=42)

# Normalizar los datos de entrenamiento
train_norm = (train - train.mean()) / train.std()
test_norm = (test - test.mean()) / test.std()

from sklearn.linear_model import LinearRegression

# Definir el modelo de regresión múltiple
regressor = LinearRegression()

# Variables predictoras
train_x = train_norm[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']]
# Variable a predecir
train_y = train_norm['CO2EMISSIONS']

# Entrenar el modelo
regressor.fit(train_x, train_y)

# Imprimir los coeficientes y la intersección
print('Coefficiente:', regressor.coef_)
print('Intercepción:', regressor.intercept_)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Variables predictoras
test_x = test_norm[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']]
# Variable a predecir
test_y = test_norm['CO2EMISSIONS']

# Realizar la predicción
test_pred = regressor.predict(test_x)

# Calcular los errores de la predicción
mae = mean_absolute_error(test_y, test_pred)
mse = mean_squared_error(test_y, test_pred)
r2 = r2_score(test_y, test_pred)

# Imprimir los errores de la predicción
print('Error absoluto:', mae)
print('Error cuadrado:', mse)
print('R2 Score:', r2)

# Graficar los datos de entrenamiento y la regresión lineal
plt.scatter(train_norm['ENGINESIZE'], train_norm['CO2EMISSIONS'], color='blue')
# plt.plot(train_norm['ENGINESIZE'], regressor.predict(train_x), color='red')
plt.scatter(train_norm['ENGINESIZE'], regressor.predict(train_x), color='red')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.show()
```
