## Problema 1:
Desarrolle una funci´on la cual reciba dos par´ametros, una imagen y un entero llamado color, la
funci´on debe devolver una imagen la cual tenga activos los canales de color segu´n los siguientes
puntos:
Si el par´ametro color vale 1, la imagen debe mostrar activos u´nicamente el color azul.
Si el par´ametro color vale 2, la imagen debe mostrar activos u´nicamente el color verde.
Si el par´ametro color vale 3, la imagen debe mostrar activos u´nicamente el color rojo.
Si el par´ametro color vale 10, la imagen debe mostrar activos u´nicamente los colores rojo y
verde.
Si el par´ametro color vale 20, la imagen debe mostrar activos u´nicamente los colores verde y
azul.
Si el par´ametro color vale 30, la imagen debe mostrar activos u´nicamente los colores azul y
rojo.

```python
def activar_canales_color(imagen, color):
    # Verificar si el parámetro color es válido
    if color not in [1, 2, 3, 10, 20, 30]:
        raise ValueError("El valor del parámetro 'color' no es válido.")

    # Crear una copia de la imagen original
    imagen_activada = np.copy(imagen)

    # Desactivar todos los canales
    imagen_activada[:, :, 0] = 0  # Canal Azul
    imagen_activada[:, :, 1] = 0  # Canal Verde
    imagen_activada[:, :, 2] = 0  # Canal Rojo

    # Activar los canales según el parámetro color
    if color == 1:
        imagen_activada[:, :, 0] = imagen[:, :, 0]  # Activar Canal Azul
    elif color == 2:
        imagen_activada[:, :, 1] = imagen[:, :, 1]  # Activar Canal Verde
    elif color == 3:
        imagen_activada[:, :, 2] = imagen[:, :, 2]  # Activar Canal Rojo
    elif color == 10:
        imagen_activada[:, :, 0] = imagen[:, :, 0]  # Activar Canal Azul
        imagen_activada[:, :, 1] = imagen[:, :, 1]  # Activar Canal Verde
    elif color == 20:
        imagen_activada[:, :, 1] = imagen[:, :, 1]  # Activar Canal Verde
        imagen_activada[:, :, 2] = imagen[:, :, 2]  # Activar Canal Rojo
    elif color == 30:
        imagen_activada[:, :, 0] = imagen[:, :, 0]  # Activar Canal Azul
        imagen_activada[:, :, 2] = imagen[:, :, 2]  # Activar Canal Rojo

    return imagen_activada

# Ejemplo de uso:
# Cargar la imagen con cv2
ruta_imagen = "imagen1.jpg"
imagen = cv2.imread(ruta_imagen)

# Activar los canales de color (por ejemplo, para activar solo el canal verde)
imagen_activada = activar_canales_color(imagen, 2)

# Mostrar la imagen original y la imagen activada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title("Imagen Original")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen_activada, cv2.COLOR_BGR2RGB))
plt.title("Imagen Activada (Canal Verde)")
plt.show()
```

## Problema 3:
Cree una funci´on que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su funci´on debe crear 3 im´agenes como salida. Para entregar este ejercicio debe incluir
una las im´agenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2

```python
def crear_escala_grises_tres_dimensiones(imagen):
    # Leer la imagen a color
    img_color = cv2.imread(imagen)

    # Extraer los canales de color (BGR)
    blue_channel = img_color[:, :, 0]
    green_channel = img_color[:, :, 1]
    red_channel = img_color[:, :, 2]

    # Crear las imágenes en escala de grises para cada canal de color
    img_blue_gris = cv2.cvtColor(blue_channel, cv2.COLOR_GRAY2BGR)
    img_green_gris = cv2.cvtColor(green_channel, cv2.COLOR_GRAY2BGR)
    img_red_gris = cv2.cvtColor(red_channel, cv2.COLOR_GRAY2BGR)

    return img_blue_gris, img_green_gris, img_red_gris

# Ruta de la imagen
ruta_imagen = "imagen1.jpg"

# Crear la escala de grises en tres dimensiones
img_azul_gris, img_verde_gris, img_rojo_gris = crear_escala_grises_tres_dimensiones(ruta_imagen)

# Mostrar las imágenes
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(cv2.cvtColor(img_azul_gris, cv2.COLOR_BGR2RGB))
plt.title("Imagen Azul en Escala de Grises")

plt.subplot(132)
plt.imshow(cv2.cvtColor(img_verde_gris, cv2.COLOR_BGR2RGB))
plt.title("Imagen Verde en Escala de Grises")

plt.subplot(133)
plt.imshow(cv2.cvtColor(img_rojo_gris, cv2.COLOR_BGR2RGB))
plt.title("Imagen Rojo en Escala de Grises")

plt.tight_layout()
plt.show()
```

## Problema 4: 
Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.

```python
def calcular_histograma(imagen):
    # Convertir la imagen a escala de grises
    imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    # Calcular histogramas para cada canal de color
    hist_rojo = cv2.calcHist([imagen], [2], None, [256], [0, 256])
    hist_verde = cv2.calcHist([imagen], [1], None, [256], [0, 256])
    hist_azul = cv2.calcHist([imagen], [0], None, [256], [0, 256])
    
    # Calcular histograma para la imagen en escala de grises
    hist_gris = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])
    
    # Calcular el resumen de cada histograma
    resumen_rojo = {
        'Media': np.mean(hist_rojo),
        'Mediana': np.median(hist_rojo),
        'Máximo': np.max(hist_rojo),
        'Mínimo': np.min(hist_rojo)
    }
    
    resumen_verde = {
        'Media': np.mean(hist_verde),
        'Mediana': np.median(hist_verde),
        'Máximo': np.max(hist_verde),
        'Mínimo': np.min(hist_verde)
    }
    
    resumen_azul = {
        'Media': np.mean(hist_azul),
        'Mediana': np.median(hist_azul),
        'Máximo': np.max(hist_azul),
        'Mínimo': np.min(hist_azul)
    }
    
    resumen_gris = {
        'Media': np.mean(hist_gris),
        'Mediana': np.median(hist_gris),
        'Máximo': np.max(hist_gris),
        'Mínimo': np.min(hist_gris)
    }
    
    return resumen_rojo, resumen_verde, resumen_azul, resumen_gris

# Cargar la imagen
imagen = cv2.imread('imagen1.jpg')

# Calcular los histogramas y obtener el resumen de datos
resumen_rojo, resumen_verde, resumen_azul, resumen_gris = calcular_histograma(imagen)

# Mostrar el resumen de cada histograma
print("Resumen del Histograma Rojo:")
print(resumen_rojo)

print("\nResumen del Histograma Verde:")
print(resumen_verde)

print("\nResumen del Histograma Azul:")
print(resumen_azul)

print("\nResumen del Histograma Escala de Grises:")
print(resumen_gris)
```

```python
def mostrar_histogramas(imagen):
    # Convertir la imagen a escala de grises
    imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    # Calcular histogramas para cada canal de color
    hist_rojo = cv2.calcHist([imagen], [2], None, [256], [0, 256])
    hist_verde = cv2.calcHist([imagen], [1], None, [256], [0, 256])
    hist_azul = cv2.calcHist([imagen], [0], None, [256], [0, 256])
    
    # Calcular histograma para la imagen en escala de grises
    hist_gris = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])
    
    # Calcular la media de cada histograma
    media_rojo = np.mean(hist_rojo)
    media_verde = np.mean(hist_verde)
    media_azul = np.mean(hist_azul)
    media_gris = np.mean(hist_gris)
    
    # Crear los gráficos de histograma
    plt.figure(figsize=(10, 6))
    
    # Histograma de canal rojo
    plt.subplot(2, 2, 1)
    plt.plot(hist_rojo, color='red')
    plt.axvline(x=media_rojo, color='red', linestyle='--')
    plt.title('Histograma Rojo')
    
    # Histograma de canal verde
    plt.subplot(2, 2, 2)
    plt.plot(hist_verde, color='green')
    plt.axvline(x=media_verde, color='green', linestyle='--')
    plt.title('Histograma Verde')
    
    # Histograma de canal azul
    plt.subplot(2, 2, 3)
    plt.plot(hist_azul, color='blue')
    plt.axvline(x=media_azul, color='blue', linestyle='--')
    plt.title('Histograma Azul')
    
    # Histograma en escala de grises
    plt.subplot(2, 2, 4)
    plt.plot(hist_gris, color='black')
    plt.axvline(x=media_gris, color='black', linestyle='--')
    plt.title('Histograma Escala de Grises')
    
    plt.tight_layout()
    plt.show()

# Cargar la imagen
imagen = cv2.imread('imagen1.jpg')

# Mostrar los histogramas de la imagen
mostrar_histogramas(imagen)
```

## Problema 5:
Investigue brevemente en que consiste el espacio de color HSV y como se mapean colores a dicho especio, para entregar este ejercicio puede hacer con un archivo PDF

El espacio de color HSV (Hue, Saturation, Value) es un modelo de representación de colores que se utiliza para describir y manipular colores en términos de sus componentes de tono, saturación y brillo. A diferencia del espacio de color RGB, que representa los colores mediante la combinación de las componentes de rojo, verde y azul, el espacio de color HSV es más intuitivo para los humanos, ya que separa la información del matiz del color (tono), su intensidad (saturación) y su luminosidad (brillo).

En el espacio de color HSV:
- El componente de tono (Hue) representa el tipo de color, como rojo, verde, azul, etc. Se mide en grados y se representa en un círculo de colores.
- El componente de saturación (Saturation) indica la pureza o intensidad del color. Un valor alto significa que el color es más vívido, mientras que un valor bajo lo hace más desaturado y cercano al gris.
- El componente de brillo (Value) se refiere a la luminosidad del color. Un valor alto indica que el color es más brillante, mientras que un valor bajo lo hace más oscuro.

Para mapear colores al espacio de color HSV, primero debemos tener una representación en el espacio de color RGB. Luego, podemos convertir esta representación RGB a HSV utilizando fórmulas matemáticas específicas. El proceso de conversión implica normalmente dividir el rango de valores RGB en rangos equivalentes en el espacio HSV. Cada valor RGB se asigna a un valor correspondiente en el espacio de color HSV, lo que nos permite obtener una descripción más intuitiva de los colores basada en tono, saturación y brillo.

Esta conversión entre espacios de color es útil en aplicaciones como procesamiento de imágenes, segmentación de objetos y manipulación de colores, ya que nos permite trabajar con características específicas del color de manera más sencilla y comprensible para los humanos.
