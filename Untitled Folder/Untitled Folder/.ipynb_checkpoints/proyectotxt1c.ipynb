```python
import cv2
import numpy as np
```

```python
# Cargar la imagen
image = cv2.imread('imagen_jardin5.jpg') 
```

```python
import matplotlib.pyplot as plt
# Mostrar la imagen original
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen original')
plt.axis('off')  # Deshabilita los ejes
plt.show()
```

```python
# Preprocesamiento de la imagen 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.title('Imagen escala grises')
plt.axis('off')  # Deshabilita los ejes
plt.show()
```

```python
# Filtrado de color para resaltar la grama (importante ajustar las variables al tipo de image para mayor precisión)
lower_green = np.array([50, 50, 50])  # Umbral inferior del color verde en el espacio de color HSV 
upper_green = np.array([90, 255, 255])  # Umbral superior del color verde en el espacio de color HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, lower_green, upper_green)
```

```python
# Combinar la máscara de color con la imagen en escala de grises
gray_filtered = cv2.bitwise_and(gray, gray, mask=mask)
```

```python
plt.imshow(cv2.cvtColor(gray_filtered, cv2.COLOR_BGR2RGB))
plt.title('Imagen con mascara de color y escala grises')
plt.axis('off')  # Deshabilita los ejes
plt.show()
```

```python
# Filtrado morfológico (elimina pequeños detalles y refina los objetos)
kernel = np.ones((5, 5), np.uint8)
morphology = cv2.morphologyEx(gray_filtered, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(cv2.cvtColor(morphology, cv2.COLOR_BGR2RGB))
plt.title('Imagen con eliminación de pequeños detalles')
plt.axis('off')  # Deshabilita los ejes
plt.show()
```

```python

# Umbralización para segmentar la grama
_, thresh = cv2.threshold(morphology, 50, 255, cv2.THRESH_BINARY)
```

```python
# Eliminación de objetos pequeños
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
min_stem_area = 500  # Umbral mínimo de área para considerar un objeto como tallo de grama
filtered_contours = [contour for contour in contours if cv2.contourArea(contour) >= min_stem_area]
```

```python
# Postprocesamiento: Fusionar tallos cercanos
distance_threshold = 20  # Umbral de distancia para fusionar tallos cercanos
merged_contours = []
while len(filtered_contours) > 0:
    current_contour = filtered_contours.pop(0)
    x1, y1, w1, h1 = cv2.boundingRect(current_contour)
    merged = [current_contour]
    
    for contour in filtered_contours:
        x2, y2, w2, h2 = cv2.boundingRect(contour)
        distance = np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
        
        if distance < distance_threshold:
            merged.append(contour)
    
    # Calcular el contorno promedio de los tallos fusionados
    merged_contour = np.vstack(merged)
    merged_contours.append(merged_contour)
```

```python
# Dibujar los contornos de los tallos fusionados en la imagen original
cv2.drawContours(image, merged_contours, -1, (0, 0, 255), 2)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen con bordes en los tallos ')
plt.axis('off')  # Deshabilita los ejes
plt.show()

# Imprimir el número estimado de tallos de grama después del postprocesamiento
num_stems = len(merged_contours)
print(f'Número estimado de tallos de grama después del postprocesamiento: {num_stems}')
```

```python
# Calcular el área total cubierta por los tallos de grama
total_grass_area = 0
for contour in merged_contours:
    total_grass_area += cv2.contourArea(contour)
```

```python
# Escalar el área a metros cuadrados 

escala_conocida = True  
if escala_conocida:
    pixel_area = total_grass_area
    # 1 metro = 100 píxeles 
    metros_cuadrados = pixel_area / 10000  # Dividir por 10000 para obtener metros cuadrados
    print(f'Área total de grama estimada en metros cuadrados: {metros_cuadrados} m²')
else:
    print(f'Área total de grama estimada en píxeles cuadrados: {total_grass_area} píxeles²')
```
